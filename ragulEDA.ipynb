```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 3D contour plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.contour3D(X, Y, Z, levels=50, cmap='viridis')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Contour Plot')
plt.show()
```

```python
!pip install basemap
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Create a map
from mpl_toolkits.basemap import Basemap
m = Basemap(projection='merc', llcrnrlat=32, urcrnrlat=42, llcrnrlon=-125, urcrnrlon=-114, resolution='i')

# Draw map features
m.drawcoastlines()
m.drawcountries()
m.drawstates()
m.fillcontinents(color='lightgreen', lake_color='aqua')
m.drawmapboundary(fill_color='aqua')
# City coordinates
cities = {'Los Angeles': (34.05, -118.25),
          'San Francisco': (37.77, -122.42),
          'San Diego': (32.72, -117.16),
          'Sacramento': (38.58, -121.49)}

# Plot cities
for city, (lat, lon) in cities.items():
    x, y = m(lon, lat)
    m.plot(x, y, 'bo', markersize=8)
    plt.text(x, y, city, fontsize=12, ha='right')

plt.title('Major Cities in California')
plt.show()
import matplotlib
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np

# Create a map
m = Basemap(projection='lcc', lat_0=35, lon_0=-100, width=1E6, height=1.2E6, resolution='i')

# Draw map features
m.drawcoastlines()
m.drawcountries()
m.drawstates()
# Generate synthetic temperature data
lons = np.linspace(-110, -90, 100)
lats = np.linspace(30, 45, 100)
lons, lats = np.meshgrid(lons, lats)
temps = np.sin(lats / 5.0) * np.cos(lons / 5.0) * 20 + 10  # Synthetic temperature data

# Convert to map projection coordinates
x, y = m(lons, lats)

# Plot the temperature data
cs = m.contourf(x, y, temps, cmap='coolwarm')
m.colorbar(cs, label='Surface Temperature (C)')

plt.title('Surface Temperature')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate sample data
np.random.seed(0)
data = {
    'age': np.random.randint(18, 70, 200),
    'gender': np.random.choice(['Male', 'Female'], 200),
    'finishing_time': np.random.normal(240, 30, 200)  # finishing time in minutes
}
df = pd.DataFrame(data)
# Scatter Plot: Age vs. Finishing Time
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='finishing_time', hue='gender', palette='deep')
plt.title('Age vs. Marathon Finishing Time')
plt.xlabel('Age')
plt.ylabel('Finishing Time (minutes)')
plt.legend(title='Gender')
plt.show()
# Histogram: Distribution of Finishing Times
plt.figure(figsize=(10, 6))
sns.histplot(df['finishing_time'], bins=20, kde=True)
plt.title('Distribution of Marathon Finishing Times')
plt.xlabel('Finishing Time (minutes)')
plt.ylabel('Frequency')
plt.show()
# Histogram: Distribution of Finishing Times
plt.figure(figsize=(10, 6))
sns.histplot(df['finishing_time'], bins=20, kde=True)
plt.title('Distribution of Marathon Finishing Times')
plt.xlabel('Finishing Time (minutes)')
plt.ylabel('Frequency')
plt.show()
# Box Plot: Finishing Times by Gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='gender', y='finishing_time', palette='Set2')
plt.title('Marathon Finishing Times by Gender')
plt.xlabel('Gender')
plt.ylabel('Finishing Time (minutes)')
plt.show()
# Pair Plot: Exploring Relationships
plt.figure(figsize=(10, 6))
sns.pairplot(df, hue='gender')
plt.title('Pair Plot of Marathon Data')
plt.show()
```

```python
import pandas as pd

# Simulated dataset of working hours
data = {'Hours Worked': [40, 42, 38, 45, 50, 41, 39, 47, 44, 36]}
df = pd.DataFrame(data)

# Calculate level (mean) and spread (standard deviation)
mean = df['Hours Worked'].mean()
std_dev = df['Hours Worked'].std()

print(f"Mean Hours Worked: {mean}")
print(f"Standard Deviation of Hours Worked: {std_dev}")
```

```python
# Simulated dataset of couples' working hours
couple_data = {'Husband': [40, 42, 38, 48, 50], 'Wife': [35, 38, 32, 40, 42]}
df_couples = pd.DataFrame(couple_data)

# Summaries of couples' working hours
mean_husband = df_couples['Husband'].mean()
mean_wife = df_couples['Wife'].mean()

print(f"Average Husband's Hours: {mean_husband}")
print(f"Average Wife's Hours: {mean_wife}")
# Calculate mean, median, mode
median = df['Hours Worked'].median()
mode = df['Hours Worked'].mode()[0]

print(f"Median Hours Worked: {median}")
print(f"Mode Hours Worked: {mode}")
# Calculate range and interquartile range (IQR)
range_val = df['Hours Worked'].max() - df['Hours Worked'].min()
iqr = df['Hours Worked'].quantile(0.75) - df['Hours Worked'].quantile(0.25)

print(f"Range: {range_val}")
print(f"Interquartile Range (IQR): {iqr}")
```

```python
# Calculate quartiles
q1 = df['Hours Worked'].quantile(0.25)
q3 = df['Hours Worked'].quantile(0.75)

print(f"Q1 (25th Percentile): {q1}")
print(f"Q3 (75th Percentile): {q3}")
```

```python
# Aggregating working hours data
total_hours = df['Hours Worked'].sum()
mean_hours = df['Hours Worked'].mean()

print(f"Total Hours Worked: {total_hours}")
print(f"Mean Hours Worked: {mean_hours}")
# Simulated dataset of hours worked by age
age_data = {'Age': [22, 30, 45, 55, 60], 'Hours Worked': [35, 40, 50, 45, 38]}
df_age = pd.DataFrame(age_data)

# Visualizing the relationship between age and hours worked
import matplotlib.pyplot as plt
plt.scatter(df_age['Age'], df_age['Hours Worked'])
plt.title('Hours Worked by Age')
plt.xlabel('Age')
plt.ylabel('Hours Worked')
plt.show()
```

```python
# Simulated dataset of working hours across countries
europe_data = {'Country': ['UK', 'Germany', 'France', 'Spain', 'Italy'], 'Hours Worked': [40, 38, 35, 36, 39]}
df_europe = pd.DataFrame(europe_data)

# Bar chart comparing working hours across Europe
plt.figure(figsize=(8,6))
plt.bar(df_europe['Country'], df_europe['Hours Worked'])
plt.title('Average Weekly Working Hours Across Europe')
plt.xlabel('Country')
plt.ylabel('Hours Worked')
plt.show()
```

```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
# Display the first few rows of the dataset
print(titanic_data.head())
# Basic information about the dataset
print(titanic_data.info())
# Summary statistics
print(titanic_data.describe())
```

```python
# Handle missing values by filling or dropping
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0], inplace=True)
titanic_data.drop(columns=['Cabin'], inplace=True)  # Drop Cabin due to too many missing values

# Verify missing values
print(titanic_data.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the distribution of survival
sns.countplot(x='Survived', data=titanic_data, palette='pastel')
plt.title('Survival Distribution')
plt.show()
```

```python
# Distribution of Age
sns.histplot(titanic_data['Age'], bins=30, kde=True, color='blue')
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
# Survival rate by class
sns.barplot(x='Pclass', y='Survived', data=titanic_data, palette='viridis')
plt.title('Survival Rate by Passenger Class')
plt.show()

# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=titanic_data, palette='viridis')
plt.title('Survival Rate by Gender')
plt.show()

# Age distribution by survival
sns.boxplot(x='Survived', y='Age', data=titanic_data, palette='pastel')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Survival rate by class and gender
sns.catplot(x='Pclass', hue='Sex', col='Survived', kind='count', data=titanic_data, palette='pastel')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
corr = titanic_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
